```python
role_mapping = {
    "Java Developers/Architects Resumes": "Java Developer",
    "Web Developer Resumes": "Web Developer",
    "SQL Developers Resumes": "SQL Developer",
    "Business Analyst (BA) Resumes": "Business Analyst",
    "Network and Systems Administrators Resumes": "Network Engineer",
    "Datawarehousing, ETL, Informatica Resumes": "Data Engineer",
    "Business Intelligence, Business Object Resumes": "BI Developer",
    "Project Manager Resumes": "Project Manager",
    "Recruiter Resumes": "HR / Recruiter",
    "Data Science Resumes": "Data Scientist"
}

df["category"] = df["category"].map(role_mapping)
```

```python
def clean_job_title(title):
    title = str(title).lower()

    # data roles
    if "data scientist" in title or "machine learning" in title or "aiml" in title:
        return "Data Scientist"

    elif "data analyst" in title or "reporting analyst" in title or "bi analyst" in title or "data analysis" in title:
        return "Data Analyst"

    elif "etl" in title or "informatica" in title or "ab initio" in title:
        return "Data Engineer"

    elif "hadoop" in title or "big data" in title:
        return "Big Data Engineer"

    elif "business intelligence" in title or "bi" in title:
        return "BI Developer"


    elif "java" in title or "j2ee" in title or "javascript" in title:
        return "Java Developer"

    elif "php" in title or "laravel" in title or "codeigniter" in title:
        return "PHP Developer"

    elif (
    "frontend" in title
    or "front end" in title
    or "react" in title
    or "angular" in title
    or "javascript" in title
    or "back end" in title
    or "backend" in title
    or "ui developer" in title or "user interface"):
        return "Web Developer"

    elif "android" in title or "ios" in title or "mobile" in title or "mac" in title or "iphone" in title:
        return "Mobile Developer"

    elif "sql" in title:
        return "SQL Developer"

    elif "software engineer" in title or "software developer" in title or "application developer" in title or "senior developer" in title:
        return "Software Engineer"

    #  MANAGEMENT LAST 
    elif "manager" in title or "lead" in title:
        return "Manager / Lead"
    elif "lead" in title and "developer" not in title:
        return "Manager/Lead"
    
    # QA / TESTING
    elif "qa" in title or "quality analyst" in title or "tester" in title or "quality assurance" in title or "test engineer" in title:
        return "QA Engineer"
    # NETWORK
    elif "network" in title:
        return "Network Engineer"
    # ORACLE / DATABASE
    elif "oracle" in title or "dba" in title:
        return "Database Engineer"
    # DEVOPS
    elif "devops" in title:
        return "DevOps Engineer"
    # BUSINESS ANALYST 
    elif "business" in title or "system analyst" in title or "business analyst" in title:
        return "Business Analyst"
    # SUPPORT ROLES
    elif "support" in title or "desktop" in title or "help desk" in title:
        return "Support Engineer"
    # SAP / ERP
    elif "sap" in title or "consultant" in title:
        return "SAP Consultant"
    
     # SYSTEM / ADMIN
    elif "system administrator" in title or "systems administrator" in title or "linux" in title or "unix" in title:
        return "System Administrator"
    # DEVOPS / INFRA
    elif "infrastructure" in title or "release" in title:
        return "DevOps Engineer"
    # SUPPORT
    elif "technician" in title or "it specialist" in title:
        return "Support Engineer"
    # ARCHITECT
    elif "architect" in title:
        return "Architect"
    elif "python" in title:
        return "Python Developer"
    elif "salesforce" in title:
        return "Salesforce Developer"
    
```

```python
X = df["Text"]
y = df["clean_job_title"]
```

```python
print(X.head())
print(y.unique())
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC

pipeline = Pipeline([
    ("tfidf", TfidfVectorizer(
        max_features=5000,
        ngram_range=(1,2),
        stop_words="english"
    )),
    ("clf", SVC(kernel="linear", probability=True))
])
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    "clf__C": [0.1, 1, 10],
    "clf__kernel": ["linear", "rbf"]
}

grid = GridSearchCV(
    pipeline,
    param_grid,
    cv=3,
    scoring="accuracy",
    n_jobs=-1,
    verbose=2
)
```

```python
grid.fit(X_train, y_train)
```

```python
best_model = grid.best_estimator_

print("Best Params:", grid.best_params_)
```

```python
from sklearn.metrics import accuracy_score

y_pred = best_model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
import joblib

joblib.dump(best_model, "career_pipeline.pkl")
joblib.dump(label_encoder, "label_encoder.pkl")
```
